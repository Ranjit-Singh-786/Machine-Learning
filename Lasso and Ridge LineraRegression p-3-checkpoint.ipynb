# Lasso and Ridge algorithym

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
# from sklearn.linear_model import Lasso,Ridge         u can also this type
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
data=pd.read_csv('train.csv')
data.head()
```

```python
x=data[['MoSold','GarageArea','GarageCars','GarageYrBlt','GrLivArea','MSSubClass','1stFlrSF','2ndFlrSF','TotalBsmtSF','BsmtUnfSF','BsmtFinSF1','YearRemodAdd','OverallQual']]
y=data[['SalePrice']]
```

```python
x.head()
```

```python
y.head()
```

```python
x.isnull().sum()
```

```python
x['GarageYrBlt']=x['GarageYrBlt'].fillna(x['GarageYrBlt'].mean())
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
```

```python
print('shape of the x_train',x_train.shape)
print('shape of the x_test',x_test.shape)
print('shape of the y_train',y_train.shape)
print('shape of the y_test',y_test.shape)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_test,y_test)        # you can see my accuracy of my model
```

>>>##  and now we will use Ridge regression

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
rdg=Ridge()
```

```python
rdg.fit(x_train,y_train)
```

```python
rdg.score(x_test,y_test)   # nothing best as compared to my linearRegression
```

>>> ## and now we will use Lasso Regression

```python
ls=Lasso()
```

```python
ls.fit(x_train,y_train)
```

```python
ls.score(x_test,y_test)
# here data distribution is not in linearRegression so
# accuracy of my model is below 
```
