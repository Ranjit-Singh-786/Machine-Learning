```python
x=df3.drop('TARGET',axis=1)
y=df3[['TARGET']]
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
```

```python
x_train.shape[0]/x.shape[0]*100
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RfR=RandomForestRegressor( n_estimators=100,criterion='mse')
```

```python
RfR.fit(x_train,y_train)
```

## 😁😁😁😉

```python
RfR.score(x_test,y_test).round(4)
```

```python
int(RfR.score(x_test,y_test).round(2)*100)
# this is good accuracy for bulding the model
```

```python
y_pred1=RfR.predict(x_test)
```

# AGain Some changes in parameter

```python
RandomForest_Regressor=RandomForestRegressor( n_estimators=200,criterion='mse')
```

```python
RandomForest_Regressor.fit(x_train,y_train)
```

```python
## some changes in my accuracy
```

```python
RandomForest_Regressor.score(x_test,y_test).round(5)
```

```python
y_pred2=RandomForest_Regressor.predict(x_test)
```

```python
### MEAN SQUARE ERROR
```

```python
from sklearn.metrics import mean_squared_error
```

```python
MSE=mean_squared_error(y_test,y_pred1)
MSE                              # mean squared error
```

```python
np.sqrt(MSE)    # i think this  is overfit model
```

```python
### MSE of second model
```

```python
mse=mean_squared_error(y_test,y_pred2)
mse
```

```python
np.sqrt(mse)
```
