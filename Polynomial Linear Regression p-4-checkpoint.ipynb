```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
num_var=df.select_dtypes(include=['int64','float64'])
num_var.head()
```

```python
num_var.isnull().sum()
```

```python
drop_col=num_var.drop(columns='LotFrontage',axis=1)
```

```python
drop_col.shape
```

```python
new_df=drop_col.dropna()
```

```python
new_df.isnull().sum()
```

```python
new_df.head()      # numerical and null free data
```

# set the independent and dependent variable from the dataset

```python
x=new_df.drop(columns='SalePrice',axis=1)
y=new_df[['SalePrice']]
```

```python
x.shape,y.shape
```

# set the train and test dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=41)
```

```python
print('shape of x_train dataset :- ',x_train.shape)
print('shape of x_test dataset :- ',x_test.shape)
print('shape of y_train dataset :- ',y_train.shape)
print('shape of y_test dataset :- ',y_test.shape)
```

## feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc.fit(x_train)
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr.score(x_test,y_test)       # there is no  effect on my accuracy after using the scaling
```

# mean and root mean squared error

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)                  # actual value and predicted value
```

```python
np.sqrt(mean_squared_error(y_test,y_pred) )  # this is overfit model
```

```python
np.sqrt(mean_squared_error(y_test,y_pred) ) #there is some difference with feature scaling
# and this is overfit model
```

# and now we will apply PolynominalFeature

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2)
```

```python
poly.fit(x_train)
x_train_poly=poly.transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
x_train_poly.shape
```

```python
x_test_poly.shape
```

# i has applied the polynomialfeature , and now we will build the model

```python
lr_with_poly=LinearRegression()
```

```python
lr_with_poly.fit(x_train,y_train)
```

```python
lr_with_poly.score(x_test,y_test)   # i did not get any benefit of polynomialFeature
```

```python
y_pred2=lr_with_poly.predict(x_test)
```

```python
y_pred2
```

```python
from sklearn.metrics import mean_squared_error
```
